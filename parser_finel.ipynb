```python
# Импорт библиотек
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
# Функция парсинга второстепенных ссылок, и возвращение заголовка статьи.
def parse(link):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'
    }
    res = requests.get(link, headers=headers).text
    soup = BeautifulSoup(res, 'html.parser')
    head = soup.find_all('h1')
    mass = []
    i=0
    for item in head:
        mass.append(item.text)
    return mass[1]
```

```python
# Парсим главную страницу РТК, и получаем html-массив элементов, в которых есть ссылки.
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'
}
res = requests.get("https://spb-rtk.ru/", headers=headers).text
soup = BeautifulSoup(res, 'html.parser')
news_list = soup.find_all('a', href=True)
```

```python
# Функция вырезки элемента массива, который является ссылкой. На вход берёт html-строку, и возвращает ссылку, в виде массива из 3.
# 0-вой элемент - полная ссылка.
def clean_text(text):
    text=str(text)
    text=text.lower()
    text = re.search(r'(http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])', text)
    return text
```

```python
# Открываем датасет, куда и будем всё запихивать.
df = pd.read_excel('links.xlsx')
df
```

```python
# Итерируемся по списку html-ссылко, которые спарсили в самом начале,
# Т.к. есть ошибки при парсинге, то мы используем try except.
# Чистая ссылка = значение элемента всех ссылок, которую почистили в функции.
# Значение second = в чистой ссылке парсим хедер статьи.
# Запихиваем данные в датасет.
y=0
for i in news_list:
    try:
        clean_link = clean_text(i)
        # print(clean_link[0], mass[y])
        second = parse(clean_link[0])
        data = {'link':clean_link[0], 'first':mass[y], 'second':second}
        df = df.append(data, ignore_index=True)
    except:
        pass
    y+=1
```
