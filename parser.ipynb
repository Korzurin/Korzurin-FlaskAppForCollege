```python
def clean_link(text):
    text=str(text)
    text=text.lower()
    text = re.search(r'(http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])', text)
    return text
```

```python
def clean_text(text):
    text=str(text)
    text = re.sub('\n', '', text)
    return text
```
