```python
def clean_link(text):
    text=str(text)
    text=text.lower()
    text = re.search(r'(http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])', text)
    return text
```

```python
def clean_text(text):
    text=str(text)
    text = re.sub('\n', '', text)
    return text
```

```python
link_before = []
link_after = []
```

```python
link_after = []
years = [17,18,19,20,21]
for year in years:
    r = requests.get(f"https://spb-rtk.ru/sitemap-posttype-post.20{year}.xml")
    xml = r.text
    soup = BeautifulSoup(xml)
    sitemapTags = soup.find_all("loc")
    for link in sitemapTags:
        link = clean_link(link)
        link = link[0]
        res = requests.get(link, headers=headers).text
        soup = BeautifulSoup(res, 'html.parser')
        news_list = soup.find_all('a', href=True)
        for unused_link in news_list:
            try:
                unused_link = clean_link(unused_link)
                if unused_link[0] not in link_after and unused_link[2] == "spb-rtk.ru":
                    link_after.append(unused_link[0])
                    print(1)
                else:
                    continue
            except:
                continue
```

```python
def parse(link):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'
    }
    res = requests.get(link, headers=headers).text
    soup = BeautifulSoup(res, 'html.parser')
    head = soup.find_all('h1')
    mass = []
    i=0
    for item in head:
        mass.append(item.text)
    # content = soup.find_all('p', href=False)
    # for item_content in content:
    #     context = ''.join(str(e) for e in content)
    context = soup.find('div', class_="entry-content").text
    return mass[1], context
```

```python
for link in tqdm(range(len(link_after))):
    link = link_after[link]
    try:
        header, text = parse(link)
        if header != 'Страница не найдена!':
            text = clean_text(text)
            data = {'link':link, 'first':header, 'second':text}
            df = df.append(data, ignore_index=True)
        else:
            continue
    except:
        continue
```
